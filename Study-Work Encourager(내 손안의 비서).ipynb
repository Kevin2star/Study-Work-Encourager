```python
# 중요 사항 : 반드시 Jupyter Notebook 또는 Colab로 실행해야 한다!!(저장 기능은 이들밖에 안 됨)

# 필요한 파일 설치
try:
    import koreanize_matplotlib
    import matplotlib.pyplot as plt
    import numpy as np
except:
    !pip install koreanize-matplotlib 
    !pip install matplotlib
    !pip install numpy # 이 기능들은 오직 ipynb 파일에서만 가능!!
    import koreanize_matplotlib
    import matplotlib.pyplot as plt
    import numpy as np
import random
from datetime import datetime
import time

# 해야 할 공부 과목들/일들 입력 함수
def what_to_do_s():
    while True:
        add = input()
        print(add)
        time.sleep(1)
        if add == '여기까지':
            break
        else:
            current_study_list.append(add)
def what_to_do_w():
    while True:
        add = input()
        print(add)
        time.sleep(1)
        if add == '여기까지':
            break
        else:
            current_work_list.append(add)


# 명언 모음집
wise_saying_list = [
    "행복의 한쪽 문이 닫히면 다른 쪽 문이 열린다. 그러나 우리는 이미 닫힌 문을 오랫동안 보기 때문에 우리를 위해 열려 있는 문을 보지 못한다. - 헬렌켈러",
    "사막이 아름다운 것은 어딘가에 샘이 숨겨져 있기 때문이다. - 생택쥐페리",
    "나 자신에 대한 자신감 잃으면 온 세상이 나의 적이 된다. - 랄프 왈도 에머슨",
    "절대 어제를 후회하지 마라. 인생은 오늘의 나 안에 있고 내일은 스스로 만드는 것이다. - L.론허바드",
    "삶이 있는 한 희망은 있다. - 키케로",
    "당신이 할 수 있다고 믿든 할 수 없다고 믿든 믿는 대로 될 것이다. - 헨리 포드",
    "항상 맑으면 사막이 된다. 비가 내리고 바람이 불어야만 비옥한 땅이 된다. - 스페인 속담",
    "진정으로 웃으려면 고통을 참아야 하며, 나아가 고통을 즐길 줄 알아야 한다. - 찰리 채플린",
    "오늘은 당신의 남은 인생 중 첫 번째 날이다. - 영화 ‘아메리칸 뷰티’ 중",
    "너무 소심하고 까다롭게 자신의 행동을 고민하지 말라. 모든 인생은 실험이다. 더 많이 실험할수록 더 나아진다. - 랄프 왈도 에머슨"
]

# 메뉴
print("Study-Work Encourager")
time.sleep(1)
while True:
    start = input("\"Start\"를 입력해 주세요: ")
    if start == "Start":
        break
while True:
    print("1. 오늘의 학습(업무) 시작")
    print("2. 현재 포인트 확인")
    print("3. 중간/기말고사 결과 분석(학생들만 이용가능)")
    print("9. 초기화(최초 실행이라면 여기를 먼저 선택해 주세요)")
    print("0. 종료")
    time.sleep(1)
    select = int(input("원하는 메뉴를 선택해 주세요: "))

    # 종료
    if select == 0:
        print("안녕히 가십시오. 종료합니다.")
        break

    # 초기화
    elif select == 9:
        # 초기화 확인
        try:
            points += 1
            points -= 1
            notice = input("초기화하시겠습니까?(Y/N)(처음 사용하시는 분들께서는 Y를 입력해 주세요)")
            if notice == "Y":
                notice = input("정말로 초기화하시겠습니까? 이번이 마지막 질문이며, 초기화하면 지금까지의 모든 정보들이 다 사라지게 됩니다.(Y/N)(처음 사용하시는 분들께서는 Y를 입력해 주세요)")
                if notice == "Y":
                    print("초기화되었습니다.")
                elif notice == "N":
                    continue
                else:
                    print("잘못된 입력입니다.")
                    continue
            elif notice == "N":
                continue
            else:
                print("잘못된 입력입니다.")
                continue
        except:
            pass
        points = 0
        last_study_result = {}
        last_work_result = {}
        last_exam_subjects = {}
        last_level = "The center of the earth"
        print("Study-Work Encourager에 오신 것을 환영합니다~~")
        time.sleep(1)
        print("이 프로그램은 학생들 또는 직장인들의 학습 및 업무를 독려하며 자기주도학습을 실천할 수 있도록 만들어졌습니다.")
        time.sleep(1)

        while True:
            category = input("자기가 어떤 분류에 해당되는지 알려주시면 그에 맞춰 학습을 독려해 드립니다. 또한 기본 50포인트를 드립니다~(학생이라면 '학생', 직장인이라면 '직장'을 입력해 주세요)")
            if category == '학생':
                print("학생이시군요! 여러분의 큰 꿈을 응원합니다. 앞으로 활기차게 함께해요~")
                points += 50
                print("현재 포인트는 50포인트입니다.")
                print("오늘의 학습을 시작하세요!")
                break
            elif category == '직장':
                print("직장인이시군요. 더 나은 발전과 성장을 위해 이 프로그램이 적극적으로 도와드리겠습니다.")
                points += 50
                print("현재 포인트는 50포인트입니다.")
                print("오늘의 업무를 시작하십시오!")
                break
            else:
                print("잘못된 입력입니다. 다시 입력하시기 바랍니다.")
        continue

    # 포인트 확인
    elif select == 2:
        try:
            if category=="학생" or "직장":
                pass
        except:
            print("초기 설정이 안 되었습니다. 메뉴에서 9번을 눌러 초기 설정을 해 주세요.")
            continue
        def point_level(points): # 레벨들
            if points < 100:
                level = "The center of the earth"
            elif points < 500:
                level = "Inner core"
            elif points < 800:
                level = "Outer core"
            elif points < 1000:
                level = "Mantle"
            elif points < 1500:
                level = "Underground B3"
            elif points < 1800:
                level = "Underground B2"
            elif points < 2000:
                level = "Underground B1"
            elif points < 2500:
                level = "Ground"
            elif points < 3000:
                level = "Grass"
            elif points < 3800:
                level = "Tree"
            elif points < 5000:
                level = "Building"
            elif points < 7000:
                level = "Building Rooftop"
            elif points < 10000:
                level = "Bird"
            elif points < 14000:
                level = "Sky"
            elif points < 18000:
                level = "Earth atmosphere"
            elif points < 22000:
                level = "Moon"
            elif points < 25000:
                level = "Space"
            elif points < 30000:
                level = "Mercury"
            elif points < 35000:
                level = "Venus"
            elif points < 40000:
                level = "Mars"
            elif points < 45000:
                level = "Jupiter"
            elif points < 50000:
                level = "Saturn"
            elif points < 55000:
                level = "Uranus"
            elif points < 60000:
                level = "Neptune"
            elif points < 80000:
                level = "The Sun"
            elif points < 100000:
                level = "The Solar System"
            elif points < 150000:
                level = "Milky way Galaxy"
            elif points < 200000:
                level = "Light"
            elif points < 500000:
                level = "Black hole"
            else:
                level = "The Universe (Top level)"
            return level

        print(f"현재 포인트는 {points}포인트입니다.")
        print("당신의 현재 포인트 등급은", point_level(points), "입니다.")
        if point_level(points) != last_level: # 레벨업
            print("축하합니다~ Level UP!")
        last_level = point_level(points)
        continue

    # 학습/업무 시작(Study encourager 본 코드)
    elif select == 1:
        # 초기 설정 여부 확인
        try:
            if category=="학생" or "직장":
                pass
        except:
            print("초기 설정이 안 되었습니다. 메뉴에서 9번을 눌러 초기 설정을 해 주세요.")
            continue
        import random
        while True:
            # 지금 시간(오전 9시가 넘었다면 자동으로 시작)
            clock = datetime.now()
            clockh = clock.hour
            if clockh >= 9: # 만약 colab에서 실행할 경우, 프로그램을 시작하기 전에 clockh를 clockh+9로 바꿔야 함. colab의 시간대는 GMT를 따르고, 우리나라의 시간대는 GMT+9이기 때문.
                print(f"지금 시간은 {clockh}시입니다.")
                print("안녕하세요~")
                time.sleep(1)
                break

        if category == "학생":
            # 랜덤적으로 명언을 고르기
            ch = random.randint(0, len(wise_saying_list) - 1)
            msg = wise_saying_list[ch]
            print("오늘의 명언 :", msg)
            time.sleep(5)
            # 공부할 과목 입력하기
            print("오늘 해야 되는 과목들은 무엇인가요?(과목을 모두 입력했다면 '여기까지'라고 입력해 주세요.)")
            time.sleep(1)
            current_study_list = []
            what_to_do_s()
            if current_study_list == []: # 오늘 해야 될 학습이 없다면
                notice = input("정말로 해야 되는 과목들이 없습니까?(Y/N)(없다면 종료하게 됩니다) ")
                if notice == "Y":
                    print("오늘 해야 될 과목들이 없는 것 같네요. 그래도 공부의 끈을 놓지 마시고! 내일 만나요~")
                    print("종료합니다.")
                    break
                elif notice == "N":
                    print("학습할 과목들을 입력하세요.")
                    what_to_do_s()
                else:
                    print("잘못된 입력입니다. 학습할 과목들을 입력하세요.")
                    what_to_do_s()
            else:
                print("오늘 해야 되는 과목들은", current_study_list, "입니다. 힘차게 공부하시고! 있다가 만나요~")
                time.sleep(1)
                # 결과 분석 및 저장
                while True:
                    ok = input("학습 결과를 입력하시려면 ok를 입력해 주세요: ")
                    if ok == "ok":
                        print("오늘 열심히 공부하셨나요? 각 과목마다 공부 시간을 얼만큼 들였는지 알려주세요!")
                        study_category = []
                        study_result = []
                        current_study_result = []
                        for sub in current_study_list:
                            print(sub)
                            time.sleep(1)
                            sub_studytime=int(input(f"학습한 시간을 분 단위로 입력해 주세요(스스로 교과서를 읽거나 문제를 푸는 것 등의 자기주도학습 대상): "))
                            study_category.append("studied")
                            print("학습한 시간(분) :", sub_studytime)
                            study_result.append(sub_studytime)
                            sub_homeworktime=int(input(f"숙제한 시간을 분 단위로 입력해 주세요: "))
                            study_category.append("homework")
                            print("숙제한 시간(분) :", sub_homeworktime)
                            study_result.append(sub_homeworktime)
                            sub_videotime=int(input(f"동영상 강의를 본 시간을 분 단위로 입력해 주세요: "))
                            print("동영상 강의를 본 시간(분) :", sub_videotime)
                            study_category.append("video")
                            study_result.append(sub_videotime)
                            # 그래프 그리기
                            plt.figure(figsize=(10, 5))
                            x = np.arange(len(study_category))
                            c = ["#" + "".join([random.choice("0123456789ABCDEF") for j in range(6)]) for _ in range(len(study_category))]
                            plt.bar(x, study_result, color=c)
                            plt.xticks(x, study_category)
                            plt.xlabel("학습 구분")
                            plt.ylabel("학습 시간(분)")
                            plt.title(f"{sub} 학습 결과")
                            plt.show()
                            study_result = []
                            study_category = []
                            current_study_result.append(sub_studytime+sub_homeworktime+sub_videotime)
                        # 전날 한 학습과 비교                            
                        for i, s in enumerate(current_study_list): # enumerate를 이용해 인덱스 값 및 저장된 값에 접근.
                            t = current_study_result[i]
                            if s in last_study_result: # 만약 오늘 학습한 과목이 전날에 학습한 과목들 중에 있는 경우
                                lt = last_study_result[s] # 전날 학습한 과목의 학습 시간
                                if t > lt: # 전날보다 더 많이 공부한 경우
                                    print(f"좋았어요! {s} 과목 학습 결과, 전날보다 {t - lt}분 더 많이 하셨습니다! 10포인트 UP!")
                                    points += 10
                                elif t==lt: # 전날과 공부 시간이 같은 경우
                                    print(f"{s} 과목은 전날과 비교해 학습한 시간이 같네요. 그래도 수고하셨어요. 내일은 더욱 열심히 공부해 봐요~!")
                                else: # 전날보다 공부 시간이 적은 경우
                                    print(f"{s} 과목은 전날 대비 그리 많이 학습하지 않으셨네요. 좀 더 많은 학습이 필요하겠습니다.")
                            else: # 오늘 학습한 과목이 전날에 학습한 과목들 중에 없는 경우
                                if t != 0: # 학습한 시간이 있는 경우
                                    print(f"오! 새로운 과목 학습의 시작! 또는 전에 학습을 멈췄다가 다시 시작했나 보네요. ({s})")
                                    print("중요한 건 끈기! 학습을 멈추지 않는 여러분의 모습이 아름답습니다. 앞으로 Fighting! 10포인트 UP!")
                                    points += 10
                                else: # 학습하기로 마음먹었지만 하지 못한 경우
                                    print(f"새롭게 시작했는데(또는 다시) 성과가 별로 없네요. 학습 권장드립니다. ({s})")
                        print("오늘의 공부도 활기차게 마무리되었네요. 내일도 Study encourager와 공부 함께 해요! 안녕~")
                        print(f"현재 포인트는 {points}포인트입니다.")
                        last_study_result={s:t for s, t in zip(current_study_list, current_study_result)} # 오늘 학습한 결과를 지난 학습 결과에다가 업데이트
                        print("종료합니다.")
                        break
                    else:
                        print("잘못된 입력입니다.")
                break

        if category == "직장":
            ch = random.randint(0, len(wise_saying_list) - 1)
            msg = wise_saying_list[ch]
            print("오늘의 명언 :", msg)
            time.sleep(5)
            # 해야 할 일들 입력
            print("오늘 해야 되는 일들은 무엇인가요?(모두 입력했다면 '여기까지'를 입력하십시오)")
            current_work_list = []
            what_to_do_w()
            if not current_work_list: # 오늘 일할 것이 없을 경우
                notice = input("정말로 해야 되는 일들이 없습니까?(Y/N)(없다면 종료하게 됩니다) ")
                if notice == "예":
                    print("오늘의 일은 없는 것 같네요. 내일 만나겠습니다.")
                    break
                elif notice == "아니오":
                    print("해야 될 일들을 입력하십시오.")
                    what_to_do_w()
                else:
                    print("잘못된 입력입니다. 해야 될 일들을 입력하십시오.")
                    what_to_do_w()
            else:
                print("오늘 해야 되는 일들은", current_work_list, "입니다. 힘차게 일하십시오. 파이팅!")
                time.sleep(1)
                # 결과 분석 및 저장
                while True:
                    ok = input("성과 결과를 입력하시려면 ok를 입력해 주십시오: ")
                    if ok=="ok":
                        print("오늘 열심히 일했는지 확인해 봅시다. 각 일마다 얼만큼 시간을 들였는지 입력해 주십시오.")
                        current_work_result = []
                        for w in current_work_list:
                            print(w)
                            time.sleep(1)
                            wrst=int(input("일한 시간을 시간 단위로 입력해 주세요: "))
                            print("일한 시간(시간): ", wrst)
                            current_work_result.append(wrst)   
                        # 그래프 그리기                     
                        plt.figure(figsize=(10, 5))
                        x = np.arange(len(current_work_list))
                        c = ["#" + "".join([random.choice("0123456789ABCDEF") for j in range(6)]) for _ in range(len(current_work_list))]
                        plt.bar(x, current_work_result, color=c)
                        plt.xticks(x, current_work_list)
                        plt.xlabel("일들")
                        plt.ylabel("일한 시간(시간)")
                        plt.title("성과 결과")
                        plt.show()
                        # 전날 한 일과 비교
                        for i, w in enumerate(current_work_list):
                            t = current_work_result[i]
                            if w in last_work_result: # 오늘 한 일이 전날 한 일들 중에 있는 경우 
                                lt = last_work_result[w] # 전날 그 일을 한 시간
                                if t > lt: # 그 일을 더 오래 한 경우
                                    print(f"오! {w} 일을 전날보다 {t - lt}분 더 많이 일하셨습니다! 수고하셨습니다. 10포인트 UP!")
                                    points += 10
                                elif t==lt: # 일한 시간이 같은 경우
                                    print(f"{w} 일은 전날과 성과가 같네요. 이보다 더 성과가 향상될 수 있도록 응원합니다!")
                                else: # 일한 시간이 오늘 더 적은 경우
                                    print(f"{w} 일은 전날 대비 성과가 별로 없네요. 더 분발하셔야겠습니다.")
                            else: # 오늘 한 일이 전날 한 일들 중에 없는 경우
                                if t != 0: # 일한 시간이 있는 경우
                                    print(f"새로운 일을 시작하셨네요? ({w})")
                                    print("끈기와 열정을 놓치지 않는다면 더욱 밝은 미래는 모두의 것입니다. 앞으로 더욱 Fighting! 10포인트 UP!")
                                    points += 10
                                else: # 그 일을 하려고 했지만 못한 경우
                                    print(f"새롭게 시작했는데 성과가 별로 없네요. 더 노력하는 것이 유일한 길입니다. 그 노력을 응원합니다! ({w})")
                        print("오늘의 일도 무사히 잘 마무리되었네요. 내일도 Study-Work Encourager와 함께하며 더 열심히 일합시다! 그럼 이만...")
                        print(f"현재 포인트는 {points}포인트입니다.")
                        last_work_result={w:t for w, t in zip(current_work_list, current_work_result)} # 오늘 한 일의 결과를 전날의 결과에다가 업데이트
                        print("종료합니다.")
                        break
                    else:
                        print("잘못된 입력입니다.")
                break

    # 중간/기말고사 결과 분석
    elif select == 3:
        # 초기 설정 여부 확인
        try:
            if category=="학생" or "직장":
                pass
        except:
            print("초기 설정이 안 되었습니다. 메뉴에서 9번을 눌러 초기 설정을 해 주세요.")
            continue
        # 사용자가 학생인지 확인
        if category != '학생':
            print("이 기능은 학생만 이용 가능합니다.")
            continue
        # 과목 입력
        print("시험을 본 과목들을 입력해 주세요. 모두 입력했으면 '여기까지'라고 입력하세요.")
        time.sleep(1)
        current_exam_subjects = []
        current_exam_scores = []
        while True:
            subject = input()
            if subject == '여기까지':
                break
            current_exam_subjects.append(subject)

        if not current_exam_subjects:
            print("입력된 과목이 없습니다. 메뉴로 돌아갑니다.")
            continue
        
        # 점수 입력
        for subject in current_exam_subjects:
            score = int(input(f"{subject} 과목의 점수를 입력해 주세요: "))
            print(f"{subject} 과목의 점수 :", score)
            current_exam_scores.append(score)

        # 그래프 그리기
        plt.figure(figsize=(10, 5))
        x = np.arange(len(current_exam_subjects))
        plt.bar(x, current_exam_scores, color=["#" + "".join([random.choice("0123456789ABCDEF") for j in range(6)]) for _ in range(len(current_exam_subjects))])
        plt.xticks(x, current_exam_subjects)
        plt.xlabel("과목")
        plt.ylabel("점수(점)")
        plt.title("이번 시험 결과")
        plt.show()

        # 지난 고사 점수와 비교
        for i, subject in enumerate(current_exam_subjects):
            current_score = current_exam_scores[i]
            if subject in last_exam_subjects:
                last_score = last_exam_subjects[subject]
                if current_score > last_score:
                    print(f"축하합니다! {subject} 과목의 점수가 {last_score}점에서 {current_score}점으로 올랐습니다! 50포인트 UP!")
                    points+=50
                elif current_score == last_score:
                    print(f"{subject} 과목의 점수가 동일합니다. 그래도 열심히 노력하셨습니다!")
                else:
                    print(f"{subject} 과목의 점수가 {last_score}점에서 {current_score}점으로 떨어졌습니다. 다음 고사에서는 더 열심히 공부합시다!")
            else:
                print(f"{subject} 과목은 이번에 처음 시험을 보셨네요. 잘하셨습니다! 10포인트 UP!")
                points+=10

        # 이번 고사 과목과 점수를 지난 고사 데이터로 업데이트
        last_exam_subjects = {subject: score for subject, score in zip(current_exam_subjects, current_exam_scores)}
        print(f"현재 포인트는 {points}포인트입니다.")
        print("시험 결과 분석이 완료되었습니다. 메뉴로 돌아갑니다.")
        continue
print("Created with Python.")
print("'Study-Work Encourager' and all contents are belongings to the creator Lee Huan Hee.")
print("All rights reserved.")
```
